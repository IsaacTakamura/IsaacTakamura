# Introducción a Pandas DataFrames


Un **DataFrame** es una estructura de datos bidimensional en Pandas que se asemeja a una tabla o una hoja de cálculo en la que los datos se organizan en filas y columnas. Cada columna puede contener diferentes tipos de datos (números, cadenas, fechas, etc.), y cada fila representa una observación o registro.

En este taller, exploraremos cómo trabajar con DataFrames en Python utilizando la biblioteca Pandas, centrándonos en un conjunto de datos de Pokémon de la primera generación. Vamos a utilizar diversos métodos y atributos útiles que nos permitirán comprender y manipular eficientemente nuestros datos.

## Cargar el DataFrame
Ahora, carguemos nuestro conjunto de datos y exploremos algunas operaciones básicas.

Primero importamos la librería `pandas`:

```python
import pandas as pd
```

Cargamos el DataFrame de Pokemon:

```python
url = "https://raw.githubusercontent.com/nicolasIsmaelUTP/TallerPandas/main/Pokemon.csv"
df = pd.read_csv(url)
```

# Ordenamiento y Subconjunto en Pandas
En este taller, exploraremos técnicas de ordenamiento (sorting) y subconjunto (subsetting) en Pandas, utilizando un conjunto de datos de Pokémon de la primera generación.

## Ordenamiento (Sorting)

### Ordenamiento Simple
Para ordenar nuestro DataFrame según una columna específica, usamos el método `.sort_values()`. Por ejemplo, ordenemos el DataFrame por el total de estadísticas en orden descendente:

```python
df.sort_values(by="Total",ascending=False)
```

### Ordenamiento por Múltiples Variables
Si queremos ordenar por varias columnas, podemos pasar una lista de nombres de columnas al método `.sort_values()`:

```python
#Lista
col = ["Tipo 1", "Total"]
df.sort_values(col)
```

## Subconjunto de Datos (Subsetting)

### Subconjunto de Columnas
Para seleccionar columnas específicas, proporcionamos una lista de nombres de columnas entre corchetes:

```python
col = ["Nombre", "Tipo 1", "Tipo 2"]
df[col]
```

### Subconjunto de Filas
Podemos seleccionar filas según una condición específica. Por ejemplo, seleccionemos solo los Pokémon legendarios:

```python
condicional = df["Legendario"] == True
df[condicional]
```

### Subconjunto basado en Datos de Texto
Seleccionemos Pokémon cuyo tipo principal sea 'Fantasma':

```python
condicional = df["Tipo 1"] == "Fantasma"
df[condicional]
```

### Subconjunto basado en Múltiples Condiciones
Seleccionemos Pokémon con 100 o más puntos en velocidad y que sean de tipo 'Fuego':

```python
c1 = df["Velocidad"] >= 100
c2 = df["Tipo 1"] == "Fuego"
df[c1 & c2]
```

### Subconjunto usando .isin()
Podemos seleccionar filas donde una columna coincida con algún valor en una lista:

```python
tipos_seleccionados = ["Fantasma", "Hielo", "Tierra"]
condicion = df["Tipo 1"].isin(tipos_seleccionados)
df[condicion]
```

# Visualización de Datos con Matplotlib
La visualización de datos es esencial para comprender patrones, tendencias y distribuciones en nuestros conjuntos de datos. En este taller, aprenderemos a utilizar Matplotlib, una poderosa biblioteca de visualización en Python. Utilizaremos un conjunto de datos de Pokémon de la primera generación para demostrar diversas técnicas de visualización.

## Histogramas

### Histograma Básico
Comencemos con un histograma simple para visualizar la distribución de la estadística 'Total':

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df["Defensa"])
plt.title("Distribucion de puntos totales")
plt.xlabel("Velocidad")
plt.ylabel("Recuento")
plt.show()
```

### Personalizando el Número de Bins
Ajustemos el número de bins para obtener una visualización más detallada:

```python
plt.hist(df["Defensa"], bins=20, color = "red",edgecolor = "black")
plt.title("Distribucion de puntos totales")
plt.xlabel("Velocidad")
plt.ylabel("Recuento")
plt.show()
```

## Agrupación y Estadísticas

### Agrupación y Promedio
Usemos el método `.groupby()` para agrupar por generación y calculemos el promedio de las estadísticas:

```python
grupo_tipo = df.groupby("Tipo 1").mean()

grupo_tipo.plot(kind="bar", y="Total", color="red")
plt.title("Barras")
plt.xlabel("Tipos de pokemon")
plt.ylabel("Promedio")
plt.show
```

## Gráfico de Dispersión (Scatter Plot)

### Creación de un Scatter Plot
Ahora, exploremos el uso de un gráfico de dispersión para visualizar la relación entre las estadísticas de ataque y defensa:

### Personalización del Scatter Plot
Filtraremos los Pokémon de tipo Fuego y Agua de un DataFrame, luego crearemos un gráfico de dispersión con los valores de ataque y defensa de los Pokémon de tipo Fuego y Agua, utilizando colores diferentes (rojo y azul) y marcadores distintos, y finalmente muestraremos el gráfico con etiquetas y leyenda.








```python
pokemon_fuego = df[ df["Tipo 1"] =="Fuego" ]
pokemon_agua = df[ df["Tipo 1"] =="Agua" ]
plt.scatter(x = pokemon_fuego["Ataque"], y = pokemon_fuego["Defensa"], color="red", marker="1")
plt.scatter(x = pokemon_agua["Ataque"], y = pokemon_agua["Defensa"], color="blue", marker="*")
plt.title("Diferencia entre Tipo Agua y Fuego con respecto al Ataque y Defensa")
plt.xlabel("Ataque")
plt.ylabel("Defensa")


plt.show
```
